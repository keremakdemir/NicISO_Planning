```python
#Importing necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Defining demand parameters
seasons = ['Winter','Spring','Summer','Fall']
hour_type = ['Peak','Offpeak']
demand_type = ['Average','Maximum']

#Defining the simulation years, generation types and all cases
years = [2023,2030,2040]
gen_types = ['Coal','Natural Gas','Fuel Oil','Nuclear']
All_cases = []
for S in seasons:
    for H in hour_type:
        for D in demand_type:
            Case_name = '{}_{}_{}'.format(S,H,D)
            All_cases.append(Case_name)

#Filtering the generation data and plotting bar plots           
for year in years:
    
    try:
        Dispatchable=pd.read_excel('../Results/Results_BAU_{}_specific_reserve_margin.xlsx'.format(year),sheet_name='DispatchableGen_MWh',header=0)
        Nuclear=pd.read_excel('../Results/Results_BAU_{}_specific_reserve_margin.xlsx'.format(year),sheet_name='NuclearGen_MWh',header=0)
        generation_comparison = pd.DataFrame(np.zeros((len(All_cases),len(gen_types))),columns=gen_types)
        Dispatchable_agg = Dispatchable.groupby('Fuel Type').sum()
        Nuclear_agg = Nuclear.groupby('Fuel Type').sum()
        
        for case in All_cases:
            
            idx = All_cases.index(case)
            Dispatchable_sp = Dispatchable_agg.loc[:,case].copy()
            Nuclear_sp = Nuclear_agg.loc[:,case].copy()
            generation_comparison.loc[idx,:] = [Dispatchable_sp['Coal'],Dispatchable_sp['Natural Gas'],Dispatchable_sp['Fuel Oil'],Nuclear_sp['Nuclear']]
        
        generation_comparison.index = All_cases
        generation_comparison = generation_comparison.apply(lambda x: round(x,1))
        
        plt.rcParams.update({'font.size': 15})
        plt.rcParams['font.sans-serif'] = "Arial"
        axis_label_pad = 10
        tick_pad = 10
        fig, ax = plt.subplots(figsize=(16,16))
        generation_comparison.plot(kind='bar', stacked=True, ax=ax, color = ['darkorange','limegreen','royalblue','red','gray','sienna'])

        # ax.xaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}')),
        # ax.set_ylabel('Generation (TWh)', labelpad=axis_label_pad)
        # ax.set_title('Generation Mix Comparison \n({})'.format(sim_details[1:]),weight='bold')
        # # ax.set_xticklabels(['Historical','Simulated'],rotation=360)
        # ax.tick_params(axis='both', which='both', pad=tick_pad)
        # plt.legend(bbox_to_anchor=(1.04,0.5), loc="center left")

        #writing the values in the boxes
        for c in ax.containers:
            ax.bar_label(c, label_type='center', weight='bold')

        plt.tight_layout()
        # plt.savefig('Gen_Mix_stacked_bar_plots/Genmix_bar ({}).png'.format(sim_details[1:]), dpi=75, bbox_inches='tight')
        plt.show()
        plt.clf()
        
        
    except FileNotFoundError:
        pass
            
```
