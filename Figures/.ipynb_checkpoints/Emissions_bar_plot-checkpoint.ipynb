```python
#Importing necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

#Defining the simulation years, generation types and all cases
years = [2023,2030,2040]
        
#Defining parameters
seasons = ['Winter','Spring','Summer','Fall']
hour_type = ['Peak','Offpeak']
demand_type = ['Average','Maximum']

All_cases = []
for S in seasons:
    for H in hour_type:
        for D in demand_type:
            Case_name = '{}_{}_{}'.format(S,H,D)
            All_cases.append(Case_name)
```

```python
#Filtering the generation data and plotting bar plots           
for year in years:
    
    if year == 2023:
        
        gen_types = ['Coal','Natural Gas','Fuel Oil']
        
        try:
            
            gens=pd.read_excel('../Data/Generators_{}.xlsx'.format(year),sheet_name='Dispatchable_{}'.format(year),header=0)
            Dispatchable=pd.read_excel('../Results/Results_BAU_{}_specific_reserve_margin_with_solar.xlsx'.format(year),sheet_name='DispatchableGen_MWh',header=0)
            dispatchable_type = Dispatchable.loc[:,'Fuel Type']
            emissions_source = pd.DataFrame(np.zeros((len(All_cases),len(gen_types))),columns=gen_types)

            for case in All_cases:

                idx = All_cases.index(case)
                Dispatchable_sp = Dispatchable.loc[:,case]
                emissions_sp = gens.loc[:,'CO2eq_Average_Emission_Rate_lb/MWh']
                Emission_mult = Dispatchable_sp*emissions_sp
                Emission_mult = Emission_mult*0.000453592
                Emission_df = pd.concat([dispatchable_type,Emission_mult],axis=1)
                Emission_df.columns = ['Fuel Type','Emissions']
                Emission_df_agg = Emission_df.groupby('Fuel Type').sum()
                emissions_source.loc[idx,:] = [Emission_df_agg.loc['Coal',:].values[0],Emission_df_agg.loc['Natural Gas',:].values[0],Emission_df_agg.loc['Fuel Oil',:].values[0]]

            emissions_source.index = All_cases
            emissions_source = emissions_source.apply(lambda x: round(x,1))
        
            seasons = ['Winter','Spring','Summer','Fall']
            
            Winter_data = emissions_source.iloc[0:4,:]
            Spring_data = emissions_source.iloc[4:8,:]
            Summer_data = emissions_source.iloc[8:12,:]
            Fall_data = emissions_source.iloc[12:16,:]
            
            fig, ax = plt.subplots(2,2 , figsize=(38,30),sharey=True)
            plt.rcParams.update({'font.size': 30})
            plt.rcParams['font.sans-serif'] = "Arial"

            axis_fontsize=25
            axis_label_pad = 10
            tick_pad = 10
            
            j=-1
            k=-1
            for S in seasons:
                if S == 'Winter' or S == 'Spring':
                    j+=1
                    globals()['{}_data'.format(S)].plot(kind='bar', stacked=True, ax=ax[0,j])
                    ax[0,j].legend('',frameon=False)
                    ax[0,j].yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
                    ax[0,j].tick_params(axis='both', which='both', pad=tick_pad)
                else:
                    k+=1
                    globals()['{}_data'.format(S)].plot(kind='bar', stacked=True, ax=ax[1,k])
                    ax[1,k].legend('',frameon=False)
                    ax[1,k].yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
                    ax[1,j].tick_params(axis='both', which='both', pad=tick_pad)
                    if k==1:
                        plt.legend(bbox_to_anchor=(1.04,1.1), loc="center left", fontsize=40)
                
                    
            ax[0,0].set_ylabel('$CO_2$ Emissions (tons)', labelpad=axis_label_pad)
            ax[1,0].set_ylabel('$CO_2$ Emissions (tons)', labelpad=axis_label_pad)
            plt.suptitle('Emission Sources ({})'.format(year),weight='bold',y=0.92, fontsize=50)
            
            ax[0,0].set_xticklabels(['Winter Peak \nAverage','Winter Peak \nMaximum','Winter Offpeak \nAverage','Winter Offpeak \nMaximum'],rotation=360)
            ax[0,1].set_xticklabels(['Spring Peak \nAverage','Spring Peak \nMaximum','Spring Offpeak \nAverage','Spring Offpeak \nMaximum'],rotation=360)
            ax[1,0].set_xticklabels(['Summer Peak \nAverage','Summer Peak \nMaximum','Summer Offpeak \nAverage','Summer Offpeak \nMaximum'],rotation=360)
            ax[1,1].set_xticklabels(['Fall Peak \nAverage','Fall Peak \nMaximum','Fall Offpeak \nAverage','Fall Offpeak \nMaximum'],rotation=360)

            plt.savefig('Emissions/Emissions ({}).png'.format(year), dpi=200, bbox_inches='tight')
            plt.show()
            plt.clf()
                
        except FileNotFoundError:
            pass
        
    else:
        
        gen_types = ['Natural Gas']
    
        try:
            
            gens=pd.read_excel('../Data/Generators_{}.xlsx'.format(year),sheet_name='Dispatchable_{}'.format(year),header=0)
            Dispatchable=pd.read_excel('../Results/Results_BAU_{}_specific_reserve_margin_high.xlsx'.format(year),sheet_name='DispatchableGen_MWh',header=0)
            dispatchable_type = Dispatchable.loc[:,'Fuel Type']
            emissions_source = pd.DataFrame(np.zeros((len(All_cases),len(gen_types))),columns=gen_types)

            for case in All_cases:

                idx = All_cases.index(case)
                Dispatchable_sp = Dispatchable.loc[:,case]
                emissions_sp = gens.loc[:,'CO2eq_Average_Emission_Rate_lb/MWh']
                Emission_mult = Dispatchable_sp*emissions_sp
                Emission_mult = Emission_mult*0.000453592
                Emission_df = pd.concat([dispatchable_type,Emission_mult],axis=1)
                Emission_df.columns = ['Fuel Type','Emissions']
                Emission_df_agg = Emission_df.groupby('Fuel Type').sum()
                emissions_source.loc[idx,:] = [Emission_df_agg.loc['Natural Gas',:].values[0]]

            emissions_source.index = All_cases
            emissions_source = emissions_source.apply(lambda x: round(x,1))
        
            seasons = ['Winter','Spring','Summer','Fall']
            
            Winter_data = emissions_source.iloc[0:4,:]
            Spring_data = emissions_source.iloc[4:8,:]
            Summer_data = emissions_source.iloc[8:12,:]
            Fall_data = emissions_source.iloc[12:16,:]
            
            fig, ax = plt.subplots(2,2 , figsize=(38,30),sharey=True)
            plt.rcParams.update({'font.size': 30})
            plt.rcParams['font.sans-serif'] = "Arial"

            axis_fontsize=25
            axis_label_pad = 10
            tick_pad = 10
            
            j=-1
            k=-1
            for S in seasons:
                if S == 'Winter' or S == 'Spring':
                    j+=1
                    globals()['{}_data'.format(S)].plot(kind='bar', stacked=True, ax=ax[0,j])
                    ax[0,j].legend('',frameon=False)
                    ax[0,j].yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
                    ax[0,j].tick_params(axis='both', which='both', pad=tick_pad)
                else:
                    k+=1
                    globals()['{}_data'.format(S)].plot(kind='bar', stacked=True, ax=ax[1,k])
                    ax[1,k].legend('',frameon=False)
                    ax[1,k].yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
                    ax[1,j].tick_params(axis='both', which='both', pad=tick_pad)
                    if k==1:
                        plt.legend(bbox_to_anchor=(1.04,1.1), loc="center left", fontsize=40)
                
                    
            ax[0,0].set_ylabel('$CO_2$ Emissions (tons)', labelpad=axis_label_pad)
            ax[1,0].set_ylabel('$CO_2$ Emissions (tons)', labelpad=axis_label_pad)
            plt.suptitle('Emission Sources ({})'.format(year),weight='bold',y=0.92, fontsize=50)
            
            ax[0,0].set_xticklabels(['Winter Peak \nAverage','Winter Peak \nMaximum','Winter Offpeak \nAverage','Winter Offpeak \nMaximum'],rotation=360)
            ax[0,1].set_xticklabels(['Spring Peak \nAverage','Spring Peak \nMaximum','Spring Offpeak \nAverage','Spring Offpeak \nMaximum'],rotation=360)
            ax[1,0].set_xticklabels(['Summer Peak \nAverage','Summer Peak \nMaximum','Summer Offpeak \nAverage','Summer Offpeak \nMaximum'],rotation=360)
            ax[1,1].set_xticklabels(['Fall Peak \nAverage','Fall Peak \nMaximum','Fall Offpeak \nAverage','Fall Offpeak \nMaximum'],rotation=360)

            plt.savefig('Emissions/Emissions ({}).png'.format(year), dpi=200, bbox_inches='tight')
            plt.show()
            plt.clf()
                
        except FileNotFoundError:
            pass

        
```
